# **4. Data Preprocessing.**

# **Pre-process the Data**
**Objective : -**
* Prepare the dataset for training by:
   1. **Cleaning the data**.
   2. **Resizing images**.
   3. **Normalizing pixel values**.
   4. Applying **data augmentation**.

````python
#Remove Corrupted Images (Data Cleaning).
from PIL import Image
import os

def remove_corrupt_images(path):
    for category in os.listdir(path):
        category_path = os.path.join(path, category)
        for img_name in os.listdir(category_path):
            img_path = os.path.join(category_path, img_name)
            try:
                with Image.open(img_path) as img:
                    img.verify()
            except:
                os.remove(img_path)
                print("Removed:", img_path)

remove_corrupt_images(dataset_path)
````

````python
#Import Required Libraries.
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator

#Import required libraries for model building.
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout,GlobalAveragePooling2D
from tensorflow.keras.applications import MobileNetV2

from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import seaborn as sns
from sklearn.metrics import roc_auc_score
from tensorflow.keras.utils import to_categorical

from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
````

````python
#Define Image Size & Augmentation.
img_size = (224, 224) #Resizing images to 224x224.
batch_size = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,          #Normalize.
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2     #80-20 split.
)
````

````python
#Create Training Data.
train_generator = train_datagen.flow_from_directory(
    dataset_path,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)
````

````python
#Check Class Indices (Label Mapping).
print("Class Indices:", train_generator.class_indices)
````

````python
#Create Validation Data.
validation_generator = train_datagen.flow_from_directory(
    dataset_path,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)
````

````python
#Calculate Class Weights.
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

#Class labels from training generator.
class_labels = train_generator.classes

#Compute weights.
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(class_labels),
    y=class_labels
)

#Convert to dictionary.
class_weights = dict(enumerate(class_weights))

print("Class Weights:", class_weights)
````

````python
#Visualize Processed Images.
images, labels = next(train_generator)

plt.figure(figsize=(10,5))
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i])
    plt.axis('off')

plt.show()
````

# **5. Feature Engineering.**

````python
#Load MobileNetV2 without top layers (Pre-trained on ImageNet).
base_model = MobileNetV2(
    weights='imagenet',      #Use pre-trained ImageNet weights.
    include_top=False,
    input_shape=(224, 224, 3)
)

#Freeze the entire base model.
base_model.trainable = False

#Verify freezing.
print("Base Model Trainable Status:", base_model.trainable)
````

# **8. Model Building.**

````python
#Build the Model.
model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(2, activation='softmax')  #2 classes: O and R.
],name="Waste_Classification_Model")
````

````python
#Compile the Model.
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
````

````python
#Display model architecture.
model.summary()
````

````python
#Check Trainable Layers.
print("Trainable layers:")
for layer in model.layers:
    print(layer.name, layer.trainable)
````

# **10. Evaluation.**

````python
#Create Non-Shuffled Validation Generator for Evaluation.
evaluation_generator = train_datagen.flow_from_directory(
    dataset_path,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation',
    shuffle=False
)
````

````python
#Generate Predictions.
predictions = model.predict(evaluation_generator)

#Convert predictions to class labels.
y_pred = np.argmax(predictions, axis=1)

#True labels.
y_true = evaluation_generator.classes
````

````python
#Classification Report.

class_labels = list(evaluation_generator.class_indices.keys())

print(classification_report(y_true, y_pred, target_names=class_labels))
````

````python
#Confusion Matrix.
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_labels,
            yticklabels=class_labels)

plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()
````

````python
#ROC-AUC Score.

#Convert true labels to categorical format.
y_true_cat = to_categorical(y_true)

#Calculate ROC-AUC score.
roc_auc = roc_auc_score(y_true_cat, predictions)

print("ROC-AUC Score:", roc_auc)
````

````python
#Error Analysis - Display Misclassified Images.

evaluation_generator.reset()

images, labels = next(evaluation_generator)

pred_probs = model.predict(images)
pred_classes = np.argmax(pred_probs, axis=1)
true_classes = np.argmax(labels, axis=1)

plt.figure(figsize=(12,8))

count = 1
for i in range(len(images)):
    if pred_classes[i] != true_classes[i]:

        plt.subplot(3,3,count)
        plt.imshow(images[i])

        plt.title(f"True: {class_labels[true_classes[i]]}\nPred: {class_labels[pred_classes[i]]}")

        plt.axis('off')

        count += 1

    if count > 9:
        break

plt.tight_layout()
plt.show()
````

# **11. Hyperparameter Tuning.**

````python
#Fine-Tuning MobileNetV2 Layers.

#Unfreeze last few layers of MobileNetV2.
base_model.trainable = True

#Freeze initial layers and train only last layers.
for layer in base_model.layers[:-20]:
    layer.trainable = False

#Check trainable status.
print("Last 20 layers are trainable for fine-tuning.")
````

````python
#Recompile Model with Lower Learning Rate.
model.compile(
    optimizer=Adam(learning_rate=0.00001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
````

````python
#Train Fine-Tuned Model with Class Weights.

fine_tune_history = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=5,
    class_weight=class_weights
)
````

````python
#Visualize Fine-Tuning Performance.

import matplotlib.pyplot as plt

#Accuracy Plot.
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(fine_tune_history.history['accuracy'], label='Training Accuracy')
plt.plot(fine_tune_history.history['val_accuracy'], label='Validation Accuracy')

plt.title("Fine-Tuning Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()

#Loss Plot.
plt.subplot(1,2,2)
plt.plot(fine_tune_history.history['loss'], label='Training Loss')
plt.plot(fine_tune_history.history['val_loss'], label='Validation Loss')

plt.title("Fine-Tuning Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()

plt.show()
````

````python
#Evaluate Fine-Tuned Model.

loss, accuracy = model.evaluate(evaluation_generator)

print("Fine-Tuned Validation Loss:", loss)
print("Fine-Tuned Validation Accuracy:", accuracy)
````
